Si ricava dopodiché con procedura analoga la tipologia di evento e l'evento secondario che porta a quello principale (esempio Assist).

```python
def evento(row):
  compi1=re.compile('(Fallo)')
  compi2=re.compile('([Tt]iro)')
  compi3=re.compile('([Cc]alcio d\'angolo)')
  compi4=re.compile('(Fuorigioco)')
  compi5=re.compile('(Sostituzione)')
  compi6=re.compile('(calcio di punizione)')
  compi7=re.compile('(ammonito)')
  compi8=re.compile('(un colpo di testa)')




  sub1 = compi1.search(row["Text"])
  sub2 = compi2.search(row["Text"])
  sub3 = compi3.search(row["Text"])
  sub4 = compi4.search(row["Text"])
  sub5 = compi5.search(row["Text"])
  sub6 = compi6.search(row["Text"])
  sub7 = compi7.search(row["Text"])
  sub8 = compi8.search(row["Text"])

  if sub1:
    return 'Foul'
  elif sub2:
    return 'Shot' 
  elif sub3:
    return 'Corner'
  elif sub4:
    return 'Offside'
  elif sub5:
    return 'Substitution'
  elif sub6:
    return 'Free kick' 
  elif sub7:
    return 'Yellow Card'  
  elif sub8:
    return 'Head-shot'
```

```python
scraping['Event'] = scraping.apply(evento, axis=1)
```

```python
def evento2(row):
  compi1=re.compile('(Assist)')
  compi2=re.compile('(lancio lungo)')
  compi3=re.compile('(passaggio filtrante)')
  



  sub1 = compi1.search(row["Text"])
  sub2 = compi2.search(row["Text"])
  sub3 = compi3.search(row["Text"])

  if sub1:
    return 'Assist'
  if sub1:
    return 'Assist'
  if sub3:
    return 'Assist'
```
