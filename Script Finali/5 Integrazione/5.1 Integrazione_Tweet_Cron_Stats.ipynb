```python
cron_c = cron_c.rename(columns={"Player": "Player_C", "Squad": "Squad_C", "Event": "Event_C", "Event2": "Event2_C", "Player2": "Player2_C"})
cron_t = cron_t.rename(columns={"Player": "Player_T", "Squad": "Squad_T", "Event": "Event_T", "Event2": "Event2_T", "Player2": "Player2_T"})
```

```python
cron_c = cron_c.drop("Unnamed: 0", axis = 1)
cron_t = cron_t.drop("Unnamed: 0", axis = 1)
```

```python
Cronaca = cron_c.merge(cron_t, how = "outer", left_on = "Minute", right_on = "Minute")
Cronaca
```

Funzione per portare il minuto della cronaca a stringa con formato per il recupero del tipo "45+2".

```python
def minuto_cr(minute):
  if minute > 90:
    rec = str(minute - 90)
    min = "90+"+rec
  elif (minute >= 45) & (minute < 48):
    rec = str(minute - 45)
    min = "45+"+rec
  else:
    min = str(minute)
  return min
```

```python
Cronaca["Minute"] = Cronaca["Minute"].apply(minuto_cr)
Cronaca
```
