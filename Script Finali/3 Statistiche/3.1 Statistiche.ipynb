Importazione librerie e dataset

```python
import numpy as np
import pandas as pd
```

```python
city_pl = pd.read_csv("Pl_city.csv", sep = ";")
tot_pl = pd.read_csv("Pl_tottenham.csv", sep = ";")
```

Si inserisce nei dataset la variabile "Team" che esplicita la squadra di appartenenza per poi concatenare i dataset.

```python
city_pl["Team"] = "Manchester City"
tot_pl["Team"] = "Tottenham"
tot_pl
```

```python
players_stat = pd.concat([tot_pl, city_pl])
players_stat
```

Si controlla il tipo degli attribuiti e si corregge la colonna Player.

```python
players_stat.dtypes
```

```python
def corr_name(old_name):
  name_arr = old_name.split("\\")
  name_arr = name_arr.pop(0)
  return str(name_arr)
```

```python
players_stat["Player"] = players_stat["Player"].apply(corr_name)
players_stat
```

La colonna relativa al ruolo presenta in alcuni casi un secondo ruolo svolto durante la partita. Viene creata una nuova colonna per esplicitare primo e secondo ruolo.

```python
def ruolo1_adj(ruolo):
  lect = list(ruolo)
  ruolo1 = lect[0]+lect[1]
  return ruolo1

def ruolo2_adj(ruolo):
  lect = list(ruolo)
  if len(lect) == 2:
    ruolo2 = None
  else:
    ruolo2 = lect[2]+lect[3]
  return ruolo2
```

```python
players_stat["First_Pos"] = players_stat["Pos"].apply(ruolo1_adj)
players_stat["Sec_Pos"] = players_stat["Pos"].apply(ruolo2_adj)
players_stat
```

```python
players_stat = players_stat.drop("Pos", axis = 1)
players_stat
```

Si esporta il dataset

```python
players_stat.to_csv("stats_players.csv")
```
